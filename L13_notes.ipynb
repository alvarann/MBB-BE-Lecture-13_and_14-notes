```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import fetch_covtype
```

# Load the Data

https://archive.ics.uci.edu/dataset/31/covertype 

```python
data = fetch_covtype()

X, y = data.data, data.target

print(X.shape)
print(y.shape)
```

```python
# only use the first 10,000 rows
X_small = X[:10_000]
y_small = y[:10_000]

print(X_small.shape)
print(y_small.shape)
```

### Inspect the data

```python
y_small_series = pd.Series(y_small)

y_small_series.value_counts()
```

### Train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(X_small, y_small, test_size=0.25, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Define model and parameter grid

```python
rf_classifier = RandomForestClassifier(bootstrap=True, random_state=42)

hyperparameter_grid = {
    'n_estimators': [1, 50, 100],
    'max_depth': [None, 2, 16],
    'max_features': ['sqrt', 'log2'],
    'max_samples': [0.1, 0.5, 0.75]
}
```

### Run the grid search

```python
grid = GridSearchCV(rf_classifier, hyperparameter_grid, cv=5, scoring='accuracy')

grid.fit(X_train, y_train)
```

### Get the best parameter model

```python
grid.best_params_
```

```python
best_rf = grid.best_estimator_

best_rf
```

### Look at feature importances

```python
importances = best_rf.feature_importances_

feature_importances_df = pd.DataFrame({
    'feature': range(54),
    'importance': importances
})

feature_importances_df.sort_values(by='importance', ascending=False)
```

### Evaluate the model on the test set

```python
y_pred = best_rf.predict(X_test)

model_accuracy = accuracy_score(y_test, y_pred)

print(model_accuracy)
```

```python
report = classification_report(y_test, y_pred)

print(report)
```
